```python
df = pd.read_csv("/content/EU_Register_on_nutrition_and_health_claims.csv")
df
```

```python
ddf = df[df[" Status"] == "Non-authorised" ]
ddf = ddf.reset_index()
data = pd.DataFrame()
data["text"] = ddf[" Claim"]
data["Food"] = ddf["Nutrient substance, food or food category"]
data["Phenotype"] = ddf[" Health relationship"]
```

```python
#Inspect data
mat = data["Phenotype"].isna()
for i in range(len(mat)):
    if mat.iloc[i]:
        print(i)
```

```python
#Remove NaN and "not validated" values
data = data[:1875]
aa = data["Phenotype"] == "not validated"
count = 0
for i in range(len(data["text"])):
    if data["Phenotype"].iloc[i] == "not validated":
        print(i)
```

```python
data = data[data["Phenotype"] != "not validated"]
len(data)
```

```python
data
```

```python
import snorkel
from snorkel.labeling import labeling_function
ABSTAIN = -1
MAINTENANCE = 0
ENHANCEMENT = 1
REDUCTION = 2
@labeling_function()
def lf_maint(x):
    return MAINTENANCE if "maintenance" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_maint2(x):
    return MAINTENANCE if "normal" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan(x):
    return ENHANCEMENT if "enhance" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan2(x):
    return ENHANCEMENT if "improves" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan3(x):
    return ENHANCEMENT if "improvement" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan4(x):
    return ENHANCEMENT if "acceleration" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan5(x):
    return ENHANCEMENT if "increase" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_enhan6(x):
    return ENHANCEMENT if "increases" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc(x):
    return REDUCTION if "reduction" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc2(x):
    return REDUCTION if "reduces" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc3(x):
    return REDUCTION if "lower" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc4(x):
    return REDUCTION if "protection" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc5(x):
    return REDUCTION if "protects" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc6(x):
    return REDUCTION if "neutralisation" in x.text.lower() else ABSTAIN
@labeling_function()
def lf_reduc7(x):
    return REDUCTION if "neutralises" in x.text.lower() else ABSTAIN
```

```python
from snorkel.labeling.model import LabelModel
from snorkel.labeling import PandasLFApplier

lfs = [lf_maint,lf_maint2, lf_enhan, lf_enhan2, lf_enhan3, lf_enhan4, lf_enhan5, lf_enhan6, lf_reduc, lf_reduc2, lf_reduc3, lf_reduc4, lf_reduc5, lf_reduc6, lf_reduc7]

applier = PandasLFApplier(lfs)
L_train = applier.apply(data)

label_model = LabelModel(cardinality=3, verbose=True)
label_model.fit(L_train, n_epochs=500, log_freq=50, seed=123)
data["label"] = label_model.predict(L=L_train, tie_break_policy="abstain")
```

```python
data.groupby(["label"]).count()["text"]
```

```python
data = data[data["label"] != -1]
data = data.reset_index()
```

```python
data = data.drop(columns=["index"])
```
